```python
# 도전미션
 
리스트 numbers 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. 이때, numbers에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다. 단, 제거된 후 남은 수들은 numbers의 원소의 순서를 유지해야 합니다.
 
예를 들면,
numbers = [1, 1, 3, 3, 0, 1, 1] ->  [1, 3, 0, 1] 
numbers = [4, 4, 4, 3, 3] ->  [4, 3]
 
코드를 작성해 주세요
```

```python
def num_max(a, b, c):
    big = a
    if b > big:
        big = b
    if c > big:
        big = c
    return big

num = num_max(10, 12, 7)
num
```

```python
def num_min(test):
    small = test[0]
    for i in range(len(test)):
    #for i in test:
        if test[i] <= small:
            small = test[i]
    return small

testarray = [52, 273, 32, 103, 275, 24, 57, 1, 25, -2]
num_min(testarray)
```

```python
def isLeapYear(year):
    if (year % 4 == 0 and year % 100 != 0 ) or year % 400 == 0:
            return True
    else:
        return False

print(isLeapYear(2022))
print(isLeapYear(2010))
print(isLeapYear(2000))
print(isLeapYear(1900))
```

```python
def add_many(*args): # *변수갯수를 정하지 않아도 됨 튜플로 받음
    print(args)
```

```python
def add_many(*args):
    sum = 0
    for i in args:
        sum += i
    return sum

add_many(1,2,3)
```

```python
def add_mul(*arg):
    if arg[0] == 'add':
        hap = 0
        for i in range(1,len(arg)):
            hap += arg[i]
        result = hap
        return result
    elif arg[0] == 'mul':
        gop = 1
        for i in range(1,len(arg)):
            gop *= arg[i]
        result = gop
        return result
```

```python

result = add_mul('add',1,2,3,4,5)
print(result)
```

```python

result = add_mul('mul',1,2,3,4,5)
print(result)
```

```python
def add_mul1(*arg, sel):
    if sel == 'add':
        result = 0
        for i in arg:
            result += i
    elif sel == 'mul':
        result = 1
        for i in arg:
            result *= i
    else:
        print("연산자 확인!")
        result = None
    return result
```

```python
add_mul1(1,2,3, sel = 'add') # sel에다가 값을 고정
```

```python
def add_and_mul(a,b):
    return a+b, a*b #값을 하나만 나오기에 튜플로 나옴
```

```python
add_and_mul(3,4) #결과값이 튜플로 나옴
```

```python
result1, result2 = add_and_mul(3,4) #리턴값을 담을 상자를 준비해서 각각 값을 받음
print(result1, result2)
```
