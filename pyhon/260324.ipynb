## 리스트 컴프리헨션

```python
# 반복문을 사용해서 리스트를 간단하게 생성
numbers = [i**2 for i in range(1,10+1)]
numbers
```

```python
mylist = []
for i in range(1,10+1):
    #print(i)
    mylist.append(i)
mylist
```

```python
numbers = [i for i in range(1,10+1) if i % 2 ==0 ]
numbers
```

```python
marks = [90, 25, 90, 45, 80]

for i in range(len(marks)):
    if marks[i]>=80:
        print(i,"번 학생 합격")
    else:
        print(i,"번 학생 불합격")

 
```
