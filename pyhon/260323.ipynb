```python
money = True

if money:
    print('밥을 먹어요')

print('if문 종료')
```

```python
#짝수 홀수를 구분하는 프로그램

num = int(input("숫자를 입력 : "))

if num %2 ==0:
    print("짝수")
else:
    print("홀수")
```

```python
eng = 80
math = 90

if eng >= 90 and math >= 90:
    print('둘 다 성공')
elif eng >= 90 and math <90:
    print('좀 더 노력')
elif eng < 90 and math >= 90:
    print('좀 더 노력')
else:
    print('많이 노력하세요')
```

```python
# 미션 : 성적 처리프로그램
# 90점 이상 100점 미만 A 학점
# 80점 이상 90점 미만 B 학점
# 70점 이상 80점 미만 C 학점
# 60점 이상 70점 미만 D 학점
# 60점 미만 F 학점

score = int(input("점수 입력 : "))

if score >= 90:
    print('A 학점')
elif score >= 80 and score < 90:
    print('B 학점')
elif score >= 70 and score < 80:
    print('C 학점')
elif score >= 60 and score < 70:
    print('D 학점')
else:
    print('F 학점')
```

```python
# 조건부 표현식

num = 2

result = '짝수' if num % 2 == 0 else '홀수'
result
```

## 반복문

for 변수 in 이터러블:
    반복할 문장

```python
for i in range(10,-1,-1):
    print(i)
```

```python
# for 반복문과 range()함수를 사용해서 1~10까지의 합을 출력

sum = 0

for i in range(1,11,1):
    sum = sum + i
print(sum)
```

```python
# for 반복문과 range()함수를 사용해서 1~10까지의 곱을 출력

sum = 1

for i in range(1,11,1):
    sum *=  i
print(sum)
```

```python
#입력받은 단의 구구단 출력

num = int(input("숫자 입력 : "))

for i in range(1,10):
    print(num, ' x ' , i , '= ', i*num)
```
