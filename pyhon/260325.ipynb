```python
등수구하기
 
영어 점수와 수학 점수의 평균 점수를 기준으로 학생들의 등수를 매기려고 합니다. 영어 점수와 수학 점수를 담은 2차원 리스트 score가 주어질 때, 영어 점수와 수학 점수의 평균을 기준으로 매긴 등수를 담은 리스트를 return하도록 함수를 만들어주세요
 
[[80, 70], [90, 50], [40, 70], [50, 80]] -> [1, 2, 4, 3]
[[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]] -> [4, 4, 6, 2, 2, 1, 7]
 
```

```python
score1 = [[0, 70], [90, 50], [40, 70], [50, 80]]
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]

def rankscore(x):
    tem = list(map(lambda x : (x[0]+x[1])/2,x))
    rst = sorted(tem,reverse=True)
    ranks =[]
    for i in rst:
        rank = rst.index(i)+1
        ranks.append(rank)
    return ranks

print(rankscore(score1))
print(rankscore(score2))    
```

```python
score = [[0, 70], [90, 50], [40, 70], [50, 80]]
avg = [sum(s)/len(s) for s in score]
avg_sort = sorted(avg, reverse=True)
result = sorted([avg_sort.index(i)+1 for i in avg])
result
```

```python
#소수판별함수 is_prime()
mylist = [1,2,3,4,5,6,7,8,9,16,14,17,19]

def is_prime(x):
    if x < 2:
        return False
    for i in range(2,x):
        if x % i == 0:
            return False
    return True


result = list(filter(is_prime, mylist))
print(result)
#[2,3,5,7]
 
```
