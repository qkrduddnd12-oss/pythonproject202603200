```python
data = {
    '월': pd.to_datetime(['2025-01-01', '2025-02-01', '2025-03-01', '2025-04-01', '2025-05-01', '2025-06-01']),
    '방문자수': [1500, 1800, 1700, 2100, 2500, 2300]
}
df_line = pd.DataFrame(data)
 
plt.figure(figsize=(6,4)) # figsize 설정
 
# 1. 플롯 그리기
plt.plot(
    df_line['월'],
    df_line['방문자수'],
    marker='o',
    linestyle='-',
    color='dodgerblue'
)
 
# 2. 그래프 꾸미기 (ax.set_ 대신 plt. 사용)
plt.title('2025년 상반기 월별 웹사이트 방문자 수', fontsize=16)
plt.xlabel('월', fontsize=12)
plt.ylabel('방문자 수 (명)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
 
```

```python
# 바차트, 산점도, 히스토그램, 히트맵

plt.plot()
plt.bar()
plt.barh()
plt.scatter()
plt.hist()
plt.boxplot()

```

```python
fig, ax = plt.subplots()

fruits = ['사과', '블루베리', '체리', '오렌지']
counts = [40, 100, 30, 55]
bar_labels = ['초록색', '보라색', '빨간색', '오렌지']
bar_colors = ['tab:green', 'tab:purple', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('공급량')
ax.set_title('과일 공급량과 색')
ax.legend(title='과일 색')

plt.show()
```
