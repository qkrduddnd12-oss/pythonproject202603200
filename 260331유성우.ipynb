```python
def predict_best_meteor_shower_viewing(city_name):
    
    if city_name not in cities['city'].values:
        temp =  f'{city_name}에서 서비스 할 수 없습니다.'
        return temp
    latitude = cities.loc[cities['city'] == city_name, 'latitude'].iloc[0]

    constellation_list = constellations.loc[
        (constellations['latitudestart'] >= latitude) & 
        (constellations['latitudeend'] <= latitude), 
        'constellation'].tolist() 
    
    if constellation_list == []:
       result =  f'{city_name}에서는 관측할 수 없습니다.'
       return result
    else:
       result =  f'{city_name}에서 별자리 관측이 가능합니다.\n'
    
    #print(result)

    for i in constellation_list:
        meteor_shower = meteor_showers.loc[meteor_showers['radiant']==i,'name',].iloc[0]
        meteor_shower_startdate = meteor_showers.loc[meteor_showers['radiant']==i,'startdate',].iloc[0]#.to_pydatetime().strftime('%Y년 %m월 %d일')
        meteor_shower_enddate = meteor_showers.loc[meteor_showers['radiant']==i,'enddate',].iloc[0]#.to_pydatetime().strftime('%Y년 %m월 %d일')
        #print(meteor_shower, meteor_shower_startdate, meteor_shower_enddate)

        moon_phase_list = moon_phases.loc[(moon_phases['date'] >= meteor_shower_startdate) & (moon_phases['date'] <= meteor_shower_enddate)]
        moon_phase_list['percentage'].idxmin()
        best_moon_date = moon_phase_list.loc[moon_phase_list['percentage'].idxmin()]['date'].to_pydatetime().strftime('%Y년 %m월 %d일')
        result += f'{meteor_shower} 유성우를 잘보려면 {i}자리 위치를 향하여 {best_moon_date}보면 됩니다.\n'

    return result
```
