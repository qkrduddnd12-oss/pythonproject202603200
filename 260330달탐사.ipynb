```python
# 임무별로 명령 모듈 중량 차이를 구해서 CM Mass diff 만들어 missions 데이터프레임에 추가
```

```python
달모듈이름 = ['Eagle (LM-5)', 'Intrepid (LM-6)', 'Antares (LM-8)', 'Falcon (LM-10)', 'Orion (LM-11)', 'Challenger (LM-12)']
달모듈중량 = [15103, 15235, 15264, 16430, 16445, 16456]

명령모듈이름 = ['Columbia (CSM-107]', 'Yankee Clipper (CM-100])', 'Kitty Hawk (CM-110)', 'Endeavor (CM-112)', 'Casper (CM-113)', 'America (CM-114)']
명령모듈중량 = [5560, 5609, 5758, 5875, 5840, 5960]
```

```python
missions['Lunar Module(LM)']=달모듈이름
missions['LM mass(kg)']=달모듈중량
missions['LM mass diff']=missions['LM mass(kg)'].diff()
missions['Command Module(CM)']=명령모듈이름
missions['CM mass(kg)']=명령모듈중량
missions['CM mass diff']=missions['CM mass(kg)'].diff()
```

```python
missions.fillna(0, inplace=True)
```

```python
#missions.drop(columns=['Lunarmodule(LM)'], inplace=True)
```

```python
#missions.drop(columns=['CM Mass(kg)'], inplace=True)
```

```python
missions['Total weight(kg)']=missions['LM mass(kg)']+missions['CM mass(kg)']
missions['Total weight diff']=missions['LM mass diff']+missions['CM mass diff']
missions.fillna(0,inplace=True)
missions
```

```python
# 페이로드에서 승무원 영역이 차이하는 비율 : Crewed area : Payload
saturnVPayload = 43500

missions['Crewed area : Payload'] = missions['Total weight(kg)'] / saturnVPayload
missions
```

```python
# 페이로드에서 
missions['Sample : Crewed area'] =  missions['Sample weight (kg)'] / missions['Total weight(kg)']
missions['Sample : Payload'] = missions['Sample weight (kg)'] / saturnVPayload
```

```python
#artemis_mission['Sample weight(kg)'] = missions['Sample weight (kg)'].sum()
artemis_mission['Sample weight from total(kg)'] =  artemis_mission['Total weight (kg)'] * missions['Sample : Crewed area'].mean()
artemis_mission['Sample weight from payload(kg)'] = artemis_mission['Payload (kg)'] * missions['Sample : Payload'].mean()
artemis_mission['Estimated sample weight(kg)'] = (artemis_mission['Sample weight from total(kg)']+artemis_mission['Sample weight from payload(kg)']) / 2
#artemis_mission.drop(columns=['Sample weight(kg)'], inplace=True)
artemis_mission
```

## needed samples overview 만들기

```python
needed_samples = pd.DataFrame()

needed_samples['Type'] = needed_sample['Type'].unique()
needed_samples['Total weight (kg)'] = needed_sample.groupby('Type')['Weight (kg)'].sum().values
needed_samples['Average weight (kg)'] = needed_sample.groupby('Type')['Weight (kg)'].mean().values

needed_samples
```

```python
temp = df.groupby('Type')['ID'].count().reset_index()
```

```python
needed_samples = pd.merge(needed_samples, temp, on = 'Type')
needed_samples
```

```python
needed_samples.drop(columns=['ID_y'], inplace=True)
```

```python
needed_samples
```

```python
needed_samples.rename(columns={'ID_x' : 'Number of samples'}, inplace=True)
```

```python
needed_samples
```

```python
needed_samples['Percentage of rock'] = needed_samples['Number of samples'] / needed_samples['Number of samples'].sum()
needed_samples['Weight to collect'] = avg_rock * needed_samples['Percentage of rock']
needed_samples['Rocks to collect'] = needed_samples['Weight to collect'] / needed_samples['Average weight (kg)']
needed_samples
```

```python
# 보고서

import math

print("암석 채집 보고서")
for index, row in needed_samples.iterrows():
    rock_type = row['Type']
    count = math.ceil(row['Rocks to collect']) # 올림 처리
    print(f"- {rock_type}: {count}개 수집 필요")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# --- 1. 데이터 준비 (보여주신 표 데이터 입력) ---

plt.rc('font', family='Malgun Gothic')

data = {
    'Type': ['Basalt', 'Breccia', 'Crustal'],
    'Rocks to collect': [13.231731, 35.571884, 20.922917],
    'Weight to collect': [16.467267, 44.991764, 2.158103]
}
needed_samples = pd.DataFrame(data)

# 'Rocks to collect'를 정수형(올림)으로 변환한 새로운 컬럼 추가
needed_samples['Rocks_count'] = needed_samples['Rocks to collect'].apply(math.ceil)


# --- 2. 그래프 그리기 (서브플롯 사용) ---
fig, axes = plt.subplots(1, 2, figsize=(16, 6)) # 1행 2열 구조
plt.subplots_adjust(wspace=0.3) # 그래프 간격 조정

# 그래프 전체 제목
fig.suptitle('달 암석 채집 임무 목표 분석', fontsize=20, fontweight='bold', y=1.05)


# [그래프 1] 암석 종류별 채집 목표 수량 (개)
sns.barplot(x='Type', y='Rocks_count', data=needed_samples, ax=axes[0], palette='viridis')
axes[0].set_title('암석 종류별 채집 목표 수량 (개)', fontsize=16, fontweight='bold')
axes[0].set_xlabel('암석 종류 (Type)', fontsize=14)
axes[0].set_ylabel('채집 목표 수량 (개)', fontsize=14)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# 막대 위에 숫자 표시
for p in axes[0].patches:
    axes[0].annotate(f"{int(p.get_height())}개", (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=12, color='black', xytext=(0, 8),
                    textcoords='offset points')


# [그래프 2] 암석 종류별 수집 목표 무게 (kg)
sns.barplot(x='Type', y='Weight to collect', data=needed_samples, ax=axes[1], palette='magma')
axes[1].set_title('암석 종류별 수집 목표 무게 (kg)', fontsize=16, fontweight='bold')
axes[1].set_xlabel('암석 종류 (Type)', fontsize=14)
axes[1].set_ylabel('수집 목표 무게 (kg)', fontsize=14)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# 막대 위에 숫자 표시 (소수점 첫째 자리까지)
for p in axes[1].patches:
    axes[1].annotate(f"{p.get_height():.1f}kg", (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=12, color='black', xytext=(0, 8),
                    textcoords='offset points')

plt.show()
```
